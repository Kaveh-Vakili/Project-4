```python
# All created individually by Mason Schwartz

class Question:
     def __init__(self, prompt, answer):
          self.prompt = prompt
          self.answer = answer

question_prompts = [
         "1. How many libraries does UMD have on its main campus?\n(a) 6\n(b) 5\n(c) 4\n(d) 3\n\n",
         "2. Which NCAA conference is UMD in?\n(a) SEC\n(b) ACC\n(c) Big Ten\n(d) Big Twelve\n\n",
         "3. What is the team name of all UMD sports teams?\n(a) Terrapins\n(b) Bobcats\n(c) Warriors\n(d) Spartans\n\n",
         "4. What is the name of UMD's mascot?\n(a) Terry\n(b) Testudo \n(c) Thomas\n(d) Teddy\n\n",
         "5. What year was UMD founded?\n(a) 1893\n(b) 1798\n(c) 1854\n(d) 1856\n\n",
         "6. What are the official colors of UMD?\n(a) Red and Orange\n(b) Blue and Gray\n(c) Black, Red, White, and Yellow\n(d) Red, Black, Navy, and Gray\n\n",
         "7. About how many students are enrolled at UMD?\n(a) 10,000\n(b) 20,000\n(c) 40,000\n(d) 60,000\n\n",
         "8. What county is UMD located in?\n(a) Montgomery County\n(b) Prince George's County\n(c) Carroll County\n(d) Hartford County\n\n",
         "9. What is the name of UMD's newspaper?\n(a) The Diamondback\n(b) The Daily Terrapin\n(c) Testudo\n(d) The Snapping Turtle\n\n",
         "10. Which UMD school is the most competetive to get into?\n(a) A. James Clark School of Engineering\n(b) College of Arts and Humanities\n(c) College of Information Studies\n(d) Robert H. Smith School of Business\n\n"
]
    
questionsT = [
            Question(question_prompts[0], "a"),
            Question(question_prompts[1], "c"),
            Question(question_prompts[2], "a"),
            Question(question_prompts[3], "b"),
            Question(question_prompts[4], "d"),
            Question(question_prompts[5], "c"),
            Question(question_prompts[6], "c"),
            Question(question_prompts[7], "b"),
            Question(question_prompts[8], "a"),
            Question(question_prompts[9], "d"),
] 

history_prompts = [
        "1. What is the capital of Maryland?\n(a) Annapolis\n(b) Baltimore\n(c) Gaithersburg\n(d) Columbia\n\n",
        "2. Who is the Governor of Maryland?\n(a) John Carney\n(b) Larry Hogan\n(c) Brian Kemp\n(d) Mike Parsons\n\n",
        "3. Which body of water does NOT run through Maryland?\n(a) The Potomac River\n(b) The Chesapeake Bay\n(c) The Columbia River\n(d) Lake Needwood\n\n",
        "4. How many local jurisdictions (counties) are in Maryland?\n(a) 22\n(b) 16\n(c) 28\n(d) 24\n\n",
        "5. What is Maryland's state bird?\n(a) The Baltimore Oriole\n(b) The Annapolis Cardinal\n(c) The Ocean City Blue Jay\n(d) The Silver Spring Sparrow\n\n",
        "6. Which city in Maryland has the largest population?\n(a) Annapolis\n(b) Baltimore\n(c) Gaithersburg\n(d) Columbia\n\n",
        "7. How many colleges are located in Maryland?\n(a) 35\n(b) 60\n(c) 48\n(d) 55\n\n",
        "8. What is the state flower of Maryland?\n(a) Violets\n(b) Black_Eyed Susans\n(c) Pink Tulips\n(d) Chrysanthemums\n\n",
        "9. How many colors are in the Maryland state flag?\n(a) 2\n(b) 3\n(c) 4\n(d) 5\n\n",
        "10. Assategue Island is home to numerous herds of what kind of animal that roam free on the isalnd?\n(a) Cows\n(b) Horses\n(c) Buffalo\n(d) Goats\n\n"
]

questionsH = [
            Question(history_prompts[0], "a"),
            Question(history_prompts[1], "b"),
            Question(history_prompts[2], "c"),
            Question(history_prompts[3], "d"),
            Question(history_prompts[4], "a"),
            Question(history_prompts[5], "b"),
            Question(history_prompts[6], "c"),
            Question(history_prompts[7], "b"),
            Question(history_prompts[8], "c"),
            Question(history_prompts[9], "b"),
]    

def run_quizTrivia():
        wronglyanswered = []
        score = 0
        for question in questionsT:
            answer = input(f"{question.prompt} Answer Choice (A-D): ").lower()
            if answer == question.answer:
                score += 1
                print("\nCorrect!\n")
            else: 
                wronglyanswered.append(question.prompt[:2])
        print(f" \nEnd of quiz!\nFinal Score: {score}/{len(questionsT)}, {score/len(questionsT)*100}%\nIncorrectly Answered Questions: ")
        print(*wronglyanswered, sep = '\n')
        
def run_quizHistory():
        wronglyanswered2 = []
        score = 0
        for question in questionsH:
            answer = input(f"{question.prompt} Answer Choice (A-D): ").lower()
            if answer == question.answer:
                score += 1
                print("\nCorrect!\n")
            else: 
                wronglyanswered2.append(question.prompt[:2])
        print(f" \nEnd of quiz!\nFinal Score: {score}/{len(questionsH)}, {score/len(questionsH)*100}%\nIncorrectly Answered Questions: ")
        print(*wronglyanswered2, sep = '\n')

def main():
    choice = input(f"===== Welcome to GoQuiz! =====\n\nWhich quiz would you like to take?\n1. UMD Trivia\n2. Maryland History\n\nPlease Specify a Number: ")
    if choice == "1":
        run_quizTrivia()
    elif choice == "2":
        run_quizHistory()
    print("\nThank you for choosing GoQuiz!\n\n===== Application Ended =====")
main()


```
